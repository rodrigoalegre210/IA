### Regresión Logística

Situación: Eres médico y quieres predecir si un paciente tiene diabetes (1) o no (0) basado en su nivel de glucosa en sangre.

```python
import numpy as np
import matplotlib.pyplot as plt

from sklearn.linear_model import LogisticRegression

# Glucosa vs. Diabetes
glucosa = np.array([80, 90, 95, 110, 130, 150, 180]).reshape(-1, 1) # reshape para 2D
diabetes = np.array([0, 0, 0, 1, 1, 1, 1])

# Crear y entrenar modelo
modelo = LogisticRegression()
modelo.fit(glucosa, diabetes)

# Graficar
plt.scatter(glucosa, diabetes, color = 'blue', label = 'Datos reales')
plt.plot(glucosa, modelo.predict_proba(glucosa)[:, 1], color = 'red', label = 'Probabilidad de diabetes')
plt.xlabel('Glucosa en sangre (mg/dL)')
plt.ylabel('Diabetes (0 = No, 1 = Sí)')
plt.legend()

plt.show()
```

* glucosa.reshape(-1, 1): Convierte el array 1D a 2D (requerido por scikit-learn).
* predict_proba(): Devuelve probabilidades (ej: [0.2, 0.8] para clase 0 y 1)
